```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt


df = pd.read_csv('Rainfall.csv')
df.columns = df.columns.str.strip()


df['rainfall_label'] = df['rainfall'].map({'yes': 1, 'no': 0})
                                          
df = df.dropna(subset=df.columns, how='any')


feature_cols = ['pressure', 'maxtemp', 'temparature', 'mintemp', 'dewpoint',
                'humidity', 'cloud', 'sunshine', 'winddirection', 'windspeed', 'rainfall_label']


sample_df = df[feature_cols].sample(100, random_state=42)


plt.figure(figsize=(12, 8))
correlation = sample_df.corr()
sns.heatmap(correlation, annot=True, fmt=".2f", cmap='coolwarm', linewidths=0.5)
plt.title("Correlation Heatmap of Sample Weather Features")
plt.xticks(rotation=45)
plt.yticks(rotation=0)
plt.tight_layout()
plt.show()
```
